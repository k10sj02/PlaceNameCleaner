```python
import requests
from bs4 import BeautifulSoup

# URL of the page you want to scrape
url = "https://www.citypopulation.de/en/ethiopia/admin/"

# Send an HTTP request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the page content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all the relevant place names on the page
    # Look for the table that contains the place names (based on the page structure)
    table = soup.find('table', {'class': 'data'})
    rows = table.find_all('tr')

    # Loop through each row to extract the place names from the first column
    places = []
    for row in rows:
        columns = row.find_all('td')
        if len(columns) > 0:  # Ensure that the row has at least one column
            place_name = columns[0].text.strip()  # The place name is in the first column
            places.append(place_name)

    # Print the list of places
    print(places)

    # Print the total number of place names extracted
    print("\n")
    print(f"Total number of place names extracted: {len(places)}")

else:
    print("Failed to retrieve the page")
```

```python
reference_list = places

# Normalize names
reference_list = [name.lower().strip() for name in reference_list]
print(reference_list)

# Print the total number of place names extracted
print("\n")
print(f"Total number of place names extracted: {len(reference_list)}")
```

```python
# Expand aliases using base Python
expanded_list = []
for name in reference_list:
    if '[' in name:  # Check if the name contains aliases within square brackets
        # Split the main name and the alias
        main_name = name.split('[')[0].strip()
        aliases = name[name.find("[")+1:name.find("]")].split('/')
        # Add the main name and aliases to the expanded list
        expanded_list.append(main_name)
        expanded_list.extend(alias.strip() for alias in aliases)
    else:
        # If no aliases, just add the name
        expanded_list.append(name)

# Clean up the expanded list: Normalize to lowercase and strip whitespace
expanded_list = [name.lower().strip() for name in expanded_list]
print(expanded_list)
# Print the total number of place names extracted
print("\n")
print(f"Total number of place names extracted: {len(expanded_list)}")
```

```python
reference_list = list(set(expanded_list))
reference_list.sort()  # Sorts the list in place
print(reference_list)
```

```python
import pandas as pd
from rapidfuzz import process, fuzz

# Load your data
df = pd.read_excel('/Users/stann-omar.jones/Downloads/01_Ethiopia_name_consistency_check.xlsx', sheet_name='Consistency Check Raw')

# Load and normalize the reference list
reference_admin1 = reference_list  # Example reference list for regions
reference_admin2 = reference_list  # Example reference list for districts

# Normalize reference lists
reference_admin1 = [name.lower().strip() for name in reference_admin1]
reference_admin2 = [name.lower().strip() for name in reference_admin2]

# Normalize raw columns in the DataFrame
df['admin1_raw_normalized'] = df['admin1_raw'].str.lower().str.strip()
df['admin2_raw_normalized'] = df['admin2_raw'].str.lower().str.strip()

# Define a function for fuzzy matching with fallback logic
def fuzzy_match(value, reference_list, threshold=80):
    if pd.isna(value):
        return None, None
    match = process.extractOne(value, reference_list, scorer=fuzz.ratio)
    if match and match[1] >= threshold:  # Strict match
        return match[0], match[1]
    
    # Fallback: Attempt lower threshold match
    match = process.extractOne(value, reference_list, scorer=fuzz.ratio, score_cutoff=60)
    if match:
        return match[0], match[1]
    
    return None, None

# Define a function for hybrid compound name matching
def match_compound_name(value, reference_list, threshold=80):
    if pd.isna(value):
        return None, None
    parts = value.split()
    
    # Strictly match full value first
    match_full = fuzzy_match(value, reference_list, threshold)
    if match_full[0]:
        return match_full
    
    # Fallback: Match components
    corrected_parts = []
    max_score = 0
    for part in parts:
        match = fuzzy_match(part, reference_list, threshold)
        if match[0]:
            corrected_parts.append(match[0])
            max_score = max(max_score, match[1])
        else:
            # Skip unmatched parts
            continue
    if corrected_parts:
        corrected_name = ' '.join(corrected_parts)
        return corrected_name.strip(), max_score
    return None, None

# Apply improved fuzzy matching for Admin1
df[['Corrected_Admin_1', 'Admin1_Match_Score']] = df['admin1_raw_normalized'].apply(
    lambda x: pd.Series(fuzzy_match(x, reference_admin1))
)

# Apply hybrid compound matching for Admin2
df[['Corrected_Admin_2', 'Admin2_Match_Score']] = df['admin2_raw_normalized'].apply(
    lambda x: pd.Series(match_compound_name(x, reference_admin2))
)

# Flag unmatched entries for review
df['Comments'] = df.apply(
    lambda row: 'Unmatched Admin1' if pd.isna(row['Corrected_Admin_1']) else (
        'Unmatched Admin2' if pd.isna(row['Corrected_Admin_2']) else ''
    ),
    axis=1
)

# Count and print matched records
corrected_admin1_count = df['Corrected_Admin_1'].notna().sum()
corrected_admin2_count = df['Corrected_Admin_2'].notna().sum()

print(f"Records with corrected Admin1: {corrected_admin1_count}")
print(f"Records with corrected Admin2: {corrected_admin2_count}")

# Drop temporary normalized columns
df.drop(columns=['admin1_raw_normalized', 'admin2_raw_normalized'], inplace=True)

# Save the updated DataFrame
df.to_csv('updated_data.csv', index=False)
```

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('updated_data.csv')

# Function to consolidate duplicate rows based on `Corrected_Admin_2`
def consolidate_entries(group):
    # Summing the years and vaccines for the group
    n_years = group['n_years'].sum()
    n_vaccines = min(group['n_vaccines'].max(), 4)  # max n_vaccines capped at 4
    
    # Calculate total_entries as a product of n_years and n_vaccines
    total_entries = n_years * n_vaccines
    
    # Calculate percentage as fraction of max possible vaccines (4 years * 6 vaccines = 24 vaccines)
    percent = (total_entries / 24) * 100
    
    # Flag consistency based on the percent
    consistency = 'consistent' if percent == 100 else 'inconsistent'
    
    # Return a new row with consolidated values
    return pd.Series({
        'n_years': n_years,
        'n_vaccines': n_vaccines,
        'total_entries': total_entries,
        'percent': percent,
        'consistency': consistency
    })

# Group by 'Corrected_Admin_2' and manually drop the grouping column before applying the function
grouped = df.groupby('Corrected_Admin_2')

# Apply the consolidation function without the grouping column
result = grouped.apply(lambda group: consolidate_entries(group.drop(columns=['Corrected_Admin_2'], errors='ignore'))).reset_index(drop=True)

# Display the resulting DataFrame
print(result)

# Optionally, save the output to a CSV
result.to_csv('consolidated_data.csv', index=False)
```

```python
import pandas as pd

# Load the datasets
updated_data = pd.read_excel('/Users/stann-omar.jones/Downloads/updated_data.xlsx')
reviewed_data = pd.read_excel('/Users/stann-omar.jones/Downloads/reviewed_data.xlsx')

# Merge the datasets
merged_df = updated_data.merge(
    reviewed_data,
    on=['admin1_raw', 'admin2_raw'],  # Merge on raw administrative regions
    how='left'  # Keep all rows in updated_data
)

# Replace Corrected_Admin_2 with Manual_Corrections if available
merged_df['Final_Admin_2'] = merged_df.apply(
    lambda row: row['Manual_Corrections'] if pd.notnull(row['Manual_Corrections']) else row['Corrected_Admin_2_x'],
    axis=1
)

# Keep all columns and add the Final_Admin_2 column
merged_df = merged_df.rename(columns={
    'Corrected_Admin_1_x': 'Corrected_Admin_1',
    'Corrected_Admin_2_x': 'Corrected_Admin_2'
}).drop(columns=['Corrected_Admin_1_y', 'Corrected_Admin_2_y'])

# Save the consolidated dataset to a new file
merged_df.to_excel('consolidated_dataset.xlsx', index=False)
```

```python
# Load the dataset
data = pd.read_excel('consolidated_dataset.xlsx')

# Group data by Final_Admin_2 and consolidate
consolidated_data = (
    data.groupby('Final_Admin_2', as_index=False)
    .agg({
        'n_years_x': 'sum',              # Add years
        'n_vaccines_x': lambda x: min(max(x), 4),  # Cap vaccines at 4
        'Corrected_Admin_1': 'first',   # Keep one representative region
        'Parent_Admin_1_x': 'first',    # Keep representative parent region
        'Parent_Admin_2_x': 'first',    # Keep representative parent district
        'Comments_x': 'first',          # Keep one representative comment
        'Admin1_Match_Score_x': 'mean', # Average match score for Admin1
        'Admin2_Match_Score_x': 'mean'  # Average match score for Admin2
    })
    .rename(columns={
        'n_years_x': 'n_years',
        'n_vaccines_x': 'n_vaccines'
    })
)

# Calculate total_entries and percent
consolidated_data['total_entries'] = consolidated_data['n_years'] * consolidated_data['n_vaccines']
consolidated_data['percent'] = (consolidated_data['total_entries'] / 24) * 100

# Set consistent flag
consolidated_data['consistent'] = consolidated_data['percent'].apply(lambda x: 'consistent' if x == 100 else 'inconsistent')

# Save the updated dataset
consolidated_data.to_excel('final_consolidated_dataset.xlsx', index=False)
```
